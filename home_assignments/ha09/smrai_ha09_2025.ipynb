### Домашнее задание 9: обратная кинематика

Решения всех трех задач нужно сопроводить коротким видео с демонстрацией работы.

**Задача 1**

Реализуйте аналитическую обратную кинематику на трех степенях свободы манипулятора Ned в Webots. Он должен сбивать синюю доминошку. Если переставить ее в другие координаты и подать их в вашу программу, он должен сбить ее и там.

**Задача 2**

Реализуйте стабилизацию шара на платформе Стюарта с помощью PD-контроллера. Мир в Webots в формате .wbt лежит в репозитории. Нужно будет написать контроллер, который считывает из симулятора положение шара и наклоняет платформу в нужную сторону. Рекомендуемая последовательность шагов:
1) выведите обратную кинематику для двух углов наклона платформы Стюарта (крен и тангаж, они же roll и pitch)
2) запрогайте ее, убедитесь, что платформа дайствительно наклоняется так, как надо
3) найдите разумные значения коэффициентов обратной связи для определения угла наклона в зависимости от отклонения шара
4) (пункт со звездочкой) сделайте движение по квадрату
